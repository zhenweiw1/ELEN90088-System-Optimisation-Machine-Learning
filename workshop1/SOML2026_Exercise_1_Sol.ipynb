## Question 1 (decide if a polynomial is convex) (Mark: 3 + 5 + 6 + 3 + 3 = 20 points)




Consider a polynomial with $2$ variable
$$
p(x)=\sum_{i,j: i+j\leq d} c_{ij}x_1^{i}x_2^{j}
$$ where the sum is over all nonnegative integer pairs $i,j$ whose sum is less or equal to $d$. We call $d$ the *degree* of the polynomial.

To understand the notation better, consider a polynomial of degree $2$ given by $ h(x)= x_1x_2+2x_2^2+3x_1-1$. You should convince yourself that $h(x)$  can be obtained by setting $c_{11}=1, c_{02}=2, c_{10}=3, c_{00}=-1$ and all other coefficients $c_{ij}$ to $0$ in $p(x)$.

### Answer the following questions:

1. how to decide the convexity of a  first-order ($d=1$) polynomial?
2. how to decide the convexity of a  quadratic ($d=2$) polynomial?
3. how to decide the convexity of a cubic ($d=3$) polynomial?
4. extend the result in (3) to all odd-order ($d$ is odd) polynomial.
5. how to decide the convexity of a quartic ($d=4$) polynomial? What is the difference/difficulty compared to the quadratic case? In particular, try to plot a convex quartic polynomial and a concave quartic polynomial in $1$ variable.

*Hint: It can be helpful to write down the general expression of polynomials p(x) before deciding the convexity.*

*Comment: as shown in [this paper](https://web.mit.edu/~a_a_a/Public/Publications/convexity_nphard.pdf), unless NP=P, it is computationally hard to check the convexity of a polynomial with order equal or larger than $4$. So in general, it is computationaly difficult to decide if a function is convex.*

**Answer:**
**1.** (1.5 points) A first-order polynomial in two variables has the form:

$$p(x_1, x_2) = c_{10}x_1 + c_{01}x_2 + c_{00}.$$

(1.5 points) Since the Hessian matrix (the matrix of second derivatives) of a linear function is always zero, the function is trivially convex.


**2.** (2 points) A quadratic polynomial in two variables has the general form:

$$p(x_1, x_2) = c_{20}x_1^2 + c_{02}x_2^2 + c_{11}x_1x_2 + c_{10}x_1 + c_{01}x_2 + c_{00}.$$
The Hessian matrix is given by:

$$
    H = \begin{bmatrix}
    \frac{\partial^2 p}{\partial x_1^2} & \frac{\partial^2 p}{\partial x_1 \partial x_2} \\
    \frac{\partial^2 p}{\partial x_2 \partial x_1} & \frac{\partial^2 p}{\partial x_2^2}
    \end{bmatrix}
    = \begin{bmatrix}
    2c_{20} & c_{11} \\
    c_{11} & 2c_{02}
    \end{bmatrix}.
$$
 For the polynomial to be convex, the Hessian must be positive semi-definite, which requires: (a) $2c_{20} \geq 0$ and $2c_{02} \geq 0$ (the diagonal entries must be non-negative) (2 points), (b) The determinant must be non-negative:
$4c_{20}c_{02} - c_{11}^2 \geq 0$. (1 point)

**3.** (1 point) A cubic polynomial takes the form:
$$
\begin{aligned}
    p(x_1, x_2) &= \sum_{i+j \leq 3} c_{ij} x_1^i x_2^j \\
    &= c_{30}x_1^3 + c_{03}x_2^3 + c_{12}x_1x_2^2 + c_{21}x_1^2x_2 + c_{20}x_1^2 + c_{02}x_2^2 + c_{11}x_1x_2  +  c_{10}x_1 + c_{01}x_2 + c_{00}.
\end{aligned}
$$
(1 point) The Hessian matrix is given by:
$$
\begin{aligned}
    H & = \begin{bmatrix}
    \frac{\partial^2 p}{\partial x_1^2} & \frac{\partial^2 p}{\partial x_1 \partial x_2} \\
    \frac{\partial^2 p}{\partial x_2 \partial x_1} & \frac{\partial^2 p}{\partial x_2^2}
    \end{bmatrix}\\
    & = \begin{bmatrix}
    6c_{30}x_1 + 2 c_{21}x_2 + 2 c_{20} & 2c_{12}x_2 + 2 c_{21}x_1 + c_{11} \\
    2c_{12}x_2 + 2 c_{21}x_1 + c_{11}  & 6c_{03}x_2 + 2 c_{12} x_1 + 2 c_{02}
    \end{bmatrix}
\end{aligned}
$$
Assume the eigenvalues of $H$ are $\lambda_1, \lambda_2$.
Check the convex conditions by the sign of eigenvalues
$$
\begin{aligned}
  \lambda_1 + \lambda_2 = tr(H)  \\
  \lambda_1 \cdot \lambda_2 = det(H)
\end{aligned}
$$
(1 point+ 1 point) We have
$$
\begin{aligned}
  \lambda_1 \cdot \lambda_2 & = det(H) \\
  & = (6c_{30}x_1 + 2 c_{21}x_2 + 2 c_{20}) (6c_{03}x_2 + 2 c_{12} x_1 + 2 c_{02}) - (2c_{12}x_2 + 2 c_{21}x_1 + c_{11})^2.\\
  \lambda_1 + \lambda_2 & = tr(H) \\
  & = (6c_{30}x_1 + 2 c_{21}x_2 + 2 c_{20}) + (6c_{03}x_2 + 2 c_{12} x_1 + 2 c_{02}) \\
  & = (6c_{30} + 2 c_{12}) x_1 + (6c_{03} + 2 c_{21}) x_2 + (2 c_{20} +  2 c_{02})
\end{aligned}
$$
(2 points) We know that there always exist $(x_1, x_2)$ pairs so that $(6c_{30}x_1 + 2 c_{21}x_2 + 2 c_{20}) (6c_{03}x_2 + 2 c_{12} x_1 + 2 c_{02}) < 0$, thus  $\lambda_1 \cdot \lambda_2 < 0 $, i.e., $\lambda_1$ and $\lambda_2$ have different signs. The Hessian is then not positive semi-definite, indicating concavity in those regions.
Therefore, $p(x)$ with $d=3$ is neither globally convex nor concave.


**4.** (1 point) Based on the result of (3), $\frac{\partial^2 p}{\partial x_1^2}$ and $\frac{\partial^2 p}{\partial x_2^2}$ for all odd-order $d$ polynomial $p(x)$ are also odd-order, i.e.,$(d-2)$-order polynomial. (1 point) Then there always exist $(x_1, x_2)$ pairs satisfying $\frac{\partial^2 p}{\partial x_1^2} \frac{\partial^2 p}{\partial x_2^2} \leq 0$. (1 point) The Hessian is then not positive semi-definite, indicating concavity in those regions.
Therefore, $p(x)$ with odd-order is neither globally convex nor concave.

**1.** Formulate the problem of minimizing the total energy consumption of the edge device subject to the latency deadline $T$. Prove that your formulation is a valid Geometric Program in standard form.

The objective is to minimize the total energy, which is the sum of computation and transmission energy:


$$\text{Minimize} \quad E(f, t_{tx}) = c_1 f^2 + c_2 t_{tx}^{-1} \quad \text{(2 points)}$$

The main constraint is the total latency, which consists of computation time and transmission time. This must not exceed $T$:


$$t_{comp} + t_{tx} \le T \implies c_3 f^{-1} + t_{tx} \le T\quad \text{(2 points)}$$

The optimization variables $f$ and $t_{tx}$ must also be strictly positive.


$$f > 0, \quad t_{tx} > 0\quad \text{(1+1=2 points)}$$

To be a standard Geometric Program, the objective function must be a posynomial, and the inequality constraints must be in the form of a posynomial less than or equal to $1$.

* **Objective:** $c_1 f^2 + c_2 t_{tx}^{-1}$ is a sum of monomials with positive coefficients, making it a valid **posynomial**. (1 points)
* **Constraint:** We can rearrange the latency constraint by dividing both sides by $T$:

$$\frac{c_3}{T} f^{-1} + \frac{1}{T} t_{tx} \le 1\quad \text{(1 point)}$$



The left side is a sum of monomials with positive coefficients, making it a valid **posynomial**.

Since the objective is a posynomial and the constraint is a posynomial $\le 1$, this formulation strictly conforms to the rules of Geometric Programming.



**2.** In energy minimization problems, the latency constraint is typically "tight" at the optimal solution (meaning the device will use all the available time T to save energy). Assuming a tight constraint: $1/f + t_{tx} = T \implies t_{tx} = T - 1/f$.
Substituting into the objective:

$$E(f) = f^2 + \frac{2}{T - 1/f}\quad (2 \ \text{points})$$


For $t_{tx} > 0$, we need $T - 1/f > 0 \implies f > 1/T$. For $T=3$, the domain is $f > 1/3$. (1 point)

```python
import numpy as np
import matplotlib.pyplot as plt

T_val = 3.0
# Define the domain f > 1/T. We start slightly above 1/3 to avoid division by zero.
f_vals = np.linspace(1/T_val + 0.05, 3.0, 200)

# Calculate E(f)
E_vals = f_vals**2 + 2 / (T_val - 1/f_vals)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(f_vals, E_vals, label='E(f)', color='blue')
plt.axvline(x=1/T_val, color='red', linestyle='--', label='Asymptote (f = 1/T)')
plt.title('Total Energy vs. CPU Frequency (T=3)')
plt.xlabel('CPU Frequency (f)')
plt.ylabel('Total Energy E(f)')
plt.ylim(0, 15)
plt.legend()
plt.grid(True)
```

**3.** Let $c_1 = 10^{-18}$, $c_2 = 0.1$, $c_3 = 10^9$, and $T = 2.0$ seconds. Write a Python script using `cvxpy` to solve this Geometric Program and find the optimal CPU frequency $f^*$ (2 points), the optimal transmission time $t_{tx}^*$ (2 points), and the minimum total energy (2 points).


```python
# Install cvxpy if not already available
#!pip install cvxpy

import cvxpy as cp

# 1. Define the given constants
c1 = 1e-18
c2 = 0.1
c3 = 1e9
T = 2.0

# 2. Define the optimization variables
# (Must specify pos=True for Geometric Programming)
f = cp.Variable(pos=True, name="CPU_frequency")
t_tx = cp.Variable(pos=True, name="Transmission_time")

# 3. Define the Objective Function (Posynomial)
energy_comp = c1 * f**2
energy_tx = c2 * t_tx**-1
objective = cp.Minimize(energy_comp + energy_tx)

# 4. Define the Constraints (Posynomial <= 1)
# Note: cvxpy's GP framework automatically handles the <= 1 transformation
# as long as the left side is a posynomial and the right side is a monomial/constant.
constraints = [
    (c3 * f**-1) + t_tx <= T
]

# 5. Formulate and solve the problem
prob = cp.Problem(objective, constraints)

# Crucial step: We must pass gp=True to tell cvxpy to parse this as a Geometric Program
prob.solve(gp=True)

# 6. Print the results
print(f"Optimization Status: {prob.status}")
print(f"Optimal CPU frequency (f):     {f.value:.2e} Hz")
print(f"Optimal transmission time (t_tx): {t_tx.value:.4f} s")
print(f"Computation time (t_comp):     {c3 / f.value:.4f} s")
print(f"Minimum Total Energy:          {prob.value:.4f} Joules")
```

## Question 3 (Aloha communication protocol) (Mark: 3 + 15 + 2 = 20 points)

**Aloha** is a well-known random access or _MAC_ (Media/multiple Access Control) communication protocol. In this system, multiple independent transmitters send packets (called $\textit{frames}$) to a single receiver over a shared channel. In the original Aloha, each transmitter sends whenever it has data, without coordinating with others. If two or more packets overlap in time, a collision occurs and the packets are lost; the transmitters then retry after a random delay.

In a more advanced setup, called **slotted Aloha**, the nodes can only transmit at the beginning of time slots, which are kept by a global/shared clock. The operations of slotted ALOHA in each node are decribed as follows:

* If there is only one frame (namely only one active transmitter) in a slot, then the transmission is successful and the slot is said to be a successful slot.
* If two or more frames (namely two or more active transmitters) collide in a slot, then the transmission is failed and a re-transmission is considered for each frame involved in the collision. The transmitter will retransmit its frame in each subsequent slot with probability $p$ until the frame is transmitted without a collision, where $0 \leq p \leq 1$.

Assume that there are $N$ transmitters and each transmitter independently attempts to transmit a frame in each slot with probability $p$. For one slot, let $S$ denote the _success probability_ that this slot is a successful slot.

Answer the following questions:
1. Provide an expression of $S$ defined above.
2. To maximise the $S$, define the optimisation problem to find the optimal value $p$. Note that the constraint $0 \leq p \leq 1$
should be taken into consideration. Clearly identify the objective and decision variable(s). Is the objective convex or concave? Show through derivation. Find the optimality conditions for this problem.
3. When $N$ tends to infinity, what is the maximal success probability $S$?

**Answer**

**1.** For an $N$-node slotted Aloha system, the probability that a given slot is successful is the probability that one of the nodes transmits and the remaining nodes do not transmit. Clearly, the probability that any given node has a successful transmission is

$$S = p (1-p)^{N-1}.\quad \text{(3 points)}$$


**2.** Since there are $N$ nodes, then the throughput of the system is given by

$$f(p) = N p (1-p)^{N-1}.\quad \text{(1 point)}$$



Therefore, the optimization problem is formulated as
$$
\begin{aligned}
& \max_p \quad N p (1-p)^{N-1} \quad \text{(1 point)}\\
& \text{subject to}  \quad 0 \leq p \leq 1 \quad \text{(1 point)}
\end{aligned}
$$


* Decision variable: $p$, the transmission attempt probability. (1 point)

* Objective function: $f(p)$, which is non-convex. (1 point)

To see this, we caculate the second-order derivation of $f(p)$:

$$f''(p) = N^2(1−p)^{N−3} (N−1)(p-\frac{2}{N}). \quad\text{(2 points)}$$


As can be obeserved, when $p < \frac{2}{N}$, the above derivation is negative and thus $f(p)$ is concave. (1 point)
Likewise, it can be verified that $f(p)$ is convex when $p > \frac{2}{N}$. (1 point)

* Optimal solution: $p^* = \frac{1}{N}$. (2 points)

We first find the local maxima.
By differentiating $f(p)$, we get

$$f'(p) = N(1−p)^{N−2} (1-Np).\quad \text{(1 point)}$$



Letting $f'(p) = 0$ gives $p = \frac{1}{N}$ (1 point), where

$$f''(\frac{1}{N}) = N(1−\frac{1}{N})^{N−3} (2\frac{N−1}{N}-N) < 0. \quad \text{(1 point)}$$



Thus, $p = \frac{1}{N}$ is the local maxima.

For the boundary points, we get $f(0) = f(1) = 0$. (1 point)

Thus $p^* = \frac{1}{N}$ is the maxima and the maxima efficiency is $f(p^*) = (1-\frac{1}{N})^{N-1}$.(1 point)

**3.** When $N \to \infty$, we have

$$\lim_{N \to \infty} f(p^*) = \lim_{N \to \infty} (1-\frac{1}{N})^{N-1} = \frac{1}{e} \approx 0.37\quad \text{(2 points)}$$


This means that although the channel is capable of transmitting a given frame at the full channel rate of $R$ bps, in the long run, the successful throughput of this channel will be $0.37 R$ bps.

```python
import numpy as np
import matplotlib.pyplot as plt
N = 10
p = np.arange(0,1,0.01)
plt.plot(p, N*p*(1-p)**(N-1))
```

```python
import numpy as np
from scipy.optimize import minimize

N = 10000
p_min = 0
p_max = 1

def objective(p):
  return - N * p * (1 - p) ** (N - 1)

bnd = [(p_min, p_max)]

p0 = 0

res = minimize(objective, p0, bounds=bnd)

# Print results
print(res)
print("Optimal p:", res.x)
print("Maximum throughput:", res.fun)
```

## Question 4 (Geometric programming: power control in wireless network) (Mark: 3 + 7 + 6 + 4 = 20 points)

### Example: Power Control in Wireless Communication

 *Adapted from Boyd, Kim, Vandenberghe, and Hassibi,* "[A Tutorial on Geometric Programming](https://web.stanford.edu/~boyd/papers/pdf/gp_tutorial.pdf)."

The [power control problem in wireless communications](http://winlab.rutgers.edu/~narayan/PAPERS/PC%20for%20Wireless%20Data.pdf) aims to minimise the total transmitter power available across $N$ trasmitters while concurrently achieving good (or a pre-defined minimum) performance.

The technical setup is as follows. Each transmitter $i$ transmits with a power level $P_i$ bounded below and above by a minimum and maximum level. The power of the signal received from transmitter $j$ at receiver $i$ is $G_{ij} P_{j}$, where $G_{ij} > 0$ represents the path gain (often loss) from transmitter $j$ to receiver $i$. The signal power at the intended receiver $i$ is $G_{ii} P_i$, and the interference power at receiver $i$ from other transmitters is given by $\sum_{k \neq i} G_{ik}P_k$. The (background) noise power at receiver $i$ is $\sigma_i$. Thus, the _Signal to Interference and  Noise Ratio (SINR)_ of the $i$th receiver-transmitter pair is

$$ S_i = \frac{G_{ii}P_i}{\sum_{k \neq i} G_{ik}P_k + \sigma_i }. $$

The minimum SINR represents a performance lower bound for this system, $S^{\text min}$.

The resulting optimisation problem is formulated as

$$
\begin{array}{ll}
\min_{P} & \sum_{i=1}^N P_i \\
\text{subject to} & P^{min} \leq P_i \leq P^{max}, \; \forall i \\
& \dfrac{G_{ii}P_i}{\sigma_i + \sum_{k \neq i} G_{ik}P_k} \geq S^{min} , \; \forall i \\
\end{array}
$$

```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt

# Problem data
N = 10                     # number of transmitters and receivers
sigma = 0.2 * np.ones(N)  # noise power at the receiver i
p_min = 0.1 * np.ones(N)  # minimum power at the transmitter i
p_max = 5 * np.ones(N)    # maximum power at the transmitter i
s_min = 0   # threshold SINR for each receiver

# Path gain matrix
G = np.random.uniform(0.1, 0.9, (N, N))
for i in range(N):
  G[i, i] = 1

def wpc_func(s_min):
  p = cp.Variable(shape=(N,), pos=True)
  objective = cp.Minimize(cp.sum(p))

  S_p = []
  for i in range(N):
      S_p.append(cp.sum(cp.hstack(G[i, k]*p[k] for k in range(N) if i != k)))
  S = sigma + cp.hstack(S_p)
  signal_power = cp.multiply(cp.diag(G), p)
  sinr = signal_power/S
  constraints = [
      p >= p_min,
      p <= p_max,
      s_min * S <= signal_power,
  ]
  problem = cp.Problem(objective, constraints)
  problem.solve()
  return problem.value, p.value, sinr.value


p_value, P_list, SINR_list = wpc_func(s_min)
print(f"The minimum total power = {p_value}")
print(f"The transmitter powers = {P_list}")
print(f"The SINR for transmitters = {SINR_list}")
```

```python
import numpy as np
from scipy.optimize import minimize

# Define parameters
N = 10  # Number of elements
P_min = 0.1  # Minimum P value
P_max = 5  # Maximum P value
G = np.random.uniform(0.1, 0.9, (N, N))
for i in range(N):
  G[i, i] = 1
sigma = np.full(N, 0.2)  # Random sigma values
S_min = 0.25  # Minimum S threshold

# Objective function
def objective(P):
  return np.sum(P)

# Constraint function
def constraint(P, i):
  numerator = G[i, i] * P[i]
  denominator = sigma[i] + sum(G[i, k] * P[k] for k in range(N) if k != i)
  return numerator / denominator - S_min

# Bounds for P
bounds = [(P_min, P_max) for _ in range(N)]

# Nonlinear constraints
constraints = [{
    'type': 'ineq',
    'fun': lambda P, i=i: constraint(P, i)
} for i in range(N)]

# Initial guess
P0 = np.full(N, (P_min + P_max) / 2)

# Solve the optimization problem
res = minimize(objective, P0, bounds=bounds, constraints=constraints)


# Print results
print(res)
print("Optimal P:", res.x)
print("Minimum sum of P:", res.fun)
```

```python
# Compute optimal SINR values
optimal_P = res.x
optimal_SINR = np.array([
    G[i, i] * optimal_P[i] / (sigma[i] + sum(G[i, k] * optimal_P[k] for k in range(N) if k != i))
    for i in range(N)
])
print("Optimal SINR values:", optimal_SINR)
```

## Question 5 (Gradient descent for Logistic Regression) (Mark: 5 + 10 + 5 = 20 points)



### Background
Suppose we are given a binary classification dataset consisting of $n$ independent samples $S=\{(\mathbf{x}_i,y_i)\}^n_{i=1}$, where $\mathbf x_i∈\mathbb R^d$ represents the feature vector of the $i$-th sample and $y_i∈\{0,1\}$ is the corresponding label.


Logistic regression models the probability of the event $Y=1$ given the feature vector $\mathbf x$. Specifically, logistic regression assumes a linear relationship between the features $\mathbf{x}$ and the log-odds of the event $Y = 1$:

$$
\log \frac{p(y = 1|\mathbf{x})}{1 - p(y = 1|\mathbf{x})} = \mathbf{x}^\top \mathbf{w},
$$
where $\mathbf w\in \mathbb R^d$ is the weight vector.

Given the dataset, we estimate the weight vector $\mathbf w$ by minimizing the empirical risk under the log-loss (also known as cross-entropy loss):

$$
\hat{R}(\mathbf{w}) = - \frac{1}{n} \sum_{i = 1}^{n} \ell_\mathrm{log}(y_i, \mu_i(\mathbf{w})),
$$

where $\ell_\mathrm{log}(y, \mu) = y \log \mu + (1 - y) \log (1 - \mu)$.

In this exercise you will:

* derive the gradient,
* implement gradient descent from scratch,
* verify correctness by comparing with a standard library implementation.

```python
#solution
from sklearn.metrics import log_loss
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import expit
sns.set_style('darkgrid')
plt.rcParams['figure.dpi'] = 108


def risk(X, y, w):
    """Evaluate the empirical risk under the cross-entropy (logistic) loss

    Parameters
    ----------
    X : array of shape (n_samples, n_features)
        Feature matrix. The matrix must contain a constant column to
        incorporate a non-zero bias.

    y : array of shape (n_samples,)
        Response relative to X. Binary classes must be encoded as 0 and 1.

    w : array of shape (n_features,)
        Weight vector.

    Returns
    -------
    risk : float
    """
    mu = expit(X @ w)
    return log_loss(y, mu)
    #You can use any method to calculate log-loss, here is just a reference
def grad_risk(X, y, w):
    """Evaluate the gradient of the empirical risk

    Parameters
    ----------
    X : array of shape (n_samples, n_features)
        Feature matrix. The matrix must contain a constant column to
        incorporate a non-zero bias.

    y : array of shape (n_samples,)
        Response relative to X. Binary classes must be encoded as 0 and 1.

    w : array of shape (n_features,)
        Weight vector.

    Returns
    -------
    grad_w : array of shape (n_features,)
    """

    mu = expit(X @ w)
    grad_w = X.T @ (mu - y) / X.shape[0]
    return grad_w

def update_weight_gd(X, y, w, **kwargs):
    """Performs a gradient descent weight update

    Parameters
    ----------
    X : array of shape (n_samples, n_features)
        Feature matrix. The matrix must contain a constant column to
        incorporate a non-zero bias.

    y : array of shape (n_samples,)
        Response relative to X. Binary classes must be encoded as 0 and 1.

    w : array of shape (n_features,)
        Current estimate of the weight vector.

    **kwargs :
        Keyword arguments.

    Returns
    -------
    w : array of shape (n_features,)
        Updated estimate of the weights vector.
    """
    # Get learning rate from kwargs, defaulting to 1.0 if None
    eta = kwargs.get("eta", 1)
    return w - eta * grad_risk(X, y, w)
def fit_logistic(X, y, w_init=None, max_iter = 100, tol = 1e-9,
                 update_weight = update_weight_gd, **kwargs):
    """Fits a binary logistic regression model according to the given training
    data

    Parameters
    ----------
    X : array of shape (n_samples, n_features)
        Feature matrix. The matrix must contain a constant column to
        incorporate a non-zero bias.

    y : array of shape (n_samples,)
        Response relative to `X`. Binary classes must be encoded as 0 and 1.

    w_init : array of shape (n_features,) default=None
        Initial guess for the weights vector. Defaults to a vector of
        zeroes.

    max_iter : int, default=100
        Maximum number of iterations

    tol : float, default=1e-4
        Stop when the inf-norm of the gradient falls below this value.

    update_weight : callable, default=update_weight_gd
        Callable that performs a weight update. Must have signature
        (X, y, w, **kwargs). Defaults to GD.

    **kwargs :
        Keyword arguments passed to `update_weight`.

    Returns
    -------
    w_history : list of arrays of shape (n_features,)
        History of weight vectors
    """
    if w_init is None:
        # Default weight initialisation
        w_init = np.zeros(X.shape[1], dtype=float)

    # Store history of weights
    w_history = [w_init]
    w = w_init

    for t in range(max_iter):
        w = update_weight(X, y, w, **kwargs)
        w_history.append(w)

        # Check stopping criterion
        grad_inf = np.linalg.norm(grad_risk(X, y, w), ord=np.inf)
        if grad_inf <= tol:
            break

    print("Stopping after {} iterations".format(t))
    print("Inf-norm of grad is {:.4g}".format(grad_inf))

    return w_history
w_history_gd = fit_logistic(X_b, y, max_iter = 1000, tol = 1e-9, eta = 5)
print("Final risk:", risk(X_b, y, w_history_gd[-1]))
def plot_decision_boundary(X, y, w):
    """Plots the decision boundary of a logistic regression classifier defined
    by weights `w`
    """
    fig, ax = plt.subplots()
    ax.scatter(X_b[y==0,1], X_b[y==0,2], color='b', label="$y = 0$")
    ax.scatter(X_b[y==1,1], X_b[y==1,2], color='r', label="$y = 1$")
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    ax.plot(list(xlim), [-w[0]/w[2] - w[1]/w[2] * x for x in xlim], ls = "-", color="k")
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_title("Decision boundary")
    ax.legend()
    plt.show()

plot_decision_boundary(X_b, y, w_history_gd[-1])
```
