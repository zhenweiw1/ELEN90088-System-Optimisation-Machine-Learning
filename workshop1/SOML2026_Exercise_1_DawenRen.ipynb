## Question 1 (decide if a polynomial is convex) (Mark: 3 + 5 + 6 + 3 + 3 = 20 points)




Consider a polynomial with $2$ variable
$$
p(x)=\sum_{i,j: i+j\leq d} c_{ij}x_1^{i}x_2^{j}
$$ where the sum is over all nonnegative integer pairs $i,j$ whose sum is less or equal to $d$. We call $d$ the *degree* of the polynomial.

To understand the notation better, consider a polynomial of degree $2$ given by $ h(x)= x_1x_2+2x_2^2+3x_1-1$. You should convince yourself that $h(x)$  can be obtained by setting $c_{11}=1, c_{02}=2, c_{10}=3, c_{00}=-1$ and all other coefficients $c_{ij}$ to $0$ in $p(x)$.

### Answer the following questions:

1. how to decide the convexity of a  first-order ($d=1$) polynomial?
2. how to decide the convexity of a  quadratic ($d=2$) polynomial?
3. how to decide the convexity of a cubic ($d=3$) polynomial?
4. extend the result in (3) to all odd-order ($d$ is odd) polynomial.
5. how to decide the convexity of a quartic ($d=4$) polynomial? What is the difference/difficulty compared to the quadratic case? In particular, try to plot a convex quartic polynomial and a concave quartic polynomial in $1$ variable.

*Hint: It can be helpful to write down the general expression of polynomials p(x) before deciding the convexity.*

*Comment: as shown in [this paper](https://web.mit.edu/~a_a_a/Public/Publications/convexity_nphard.pdf), unless NP=P, it is computationally hard to check the convexity of a polynomial with order equal or larger than $4$. So in general, it is computationaly difficult to decide if a function is convex.*

### Answer to Question 1

---

#### 1.1 First-order ($d=1$) polynomial

A first-order polynomial in two variables is:
$$p(x) = c_{10}x_1 + c_{01}x_2 + c_{00}$$

This is an **affine function**. The Hessian matrix is the **zero matrix**:
$$\nabla^2 p(x) = \mathbf{0} \succeq 0$$

Since the zero matrix is positive semidefinite, **all first-order polynomials are convex** (and also concave). No conditions on the coefficients are needed.

---

#### 1.2 Quadratic ($d=2$) polynomial

A quadratic polynomial in two variables is:
$$p(x) = c_{20}x_1^2 + c_{11}x_1 x_2 + c_{02}x_2^2 + c_{10}x_1 + c_{01}x_2 + c_{00}$$

The Hessian is a **constant matrix**:
$$\nabla^2 p(x) = \begin{bmatrix} 2c_{20} & c_{11} \\ c_{11} & 2c_{02} \end{bmatrix}$$

The polynomial is convex if and only if $\nabla^2 p(x) \succeq 0$, which requires:
1. $c_{20} \geq 0$ and $c_{02} \geq 0$ (diagonal entries non-negative)
2. $4 c_{20} c_{02} - c_{11}^2 \geq 0$ (determinant non-negative)

---

#### 1.3 Cubic ($d=3$) polynomial

A cubic polynomial in two variables has the general form:
$$p(x) = c_{30}x_1^3 + c_{21}x_1^2 x_2 + c_{12}x_1 x_2^2 + c_{03}x_2^3 + \text{(lower-order terms)}$$

The Hessian entries are **affine (linear)** functions of $x$:
$$H_{11} = \frac{\partial^2 p}{\partial x_1^2} = 2c_{20} + 6c_{30}x_1 + 2c_{21}x_2$$
$$H_{12} = \frac{\partial^2 p}{\partial x_1 \partial x_2} = c_{11} + 2c_{21}x_1 + 2c_{12}x_2$$
$$H_{22} = \frac{\partial^2 p}{\partial x_2^2} = 2c_{02} + 2c_{12}x_1 + 6c_{03}x_2$$

For convexity, we need $H(x) \succeq 0$ for **all** $x \in \mathbb{R}^2$. Since $H_{11}$ is affine in $x$, it can only be non-negative everywhere if the coefficients of $x_1$ and $x_2$ are zero:
$$6c_{30} = 0, \quad 2c_{21} = 0 \implies c_{30} = c_{21} = 0$$

Similarly, from $H_{22} \geq 0$ for all $x$:
$$2c_{12} = 0, \quad 6c_{03} = 0 \implies c_{12} = c_{03} = 0$$

Thus **all third-order coefficients must be zero**, reducing it to a quadratic. A cubic polynomial is convex **if and only if** all cubic-order coefficients vanish and the remaining quadratic part satisfies the conditions in (1.2).

---

#### 1.4 All odd-order ($d$ odd) polynomials

The argument in (1.3) generalizes by induction. For a polynomial of odd degree $d \geq 3$:

- The Hessian $\nabla^2 p(x)$ has entries that are polynomials of degree $d-2$ (which is also odd, $\geq 1$).
- A polynomial of **odd degree** in any variable is unbounded from below and above — it takes both positive and negative values as $x \to \pm\infty$.
- Therefore, for the diagonal entries $H_{ii}(x) \geq 0$ to hold for all $x$, the highest-order terms must vanish.
- Repeating this argument, all terms of degree greater than 2 must vanish.

**Conclusion:** An odd-degree polynomial is convex **if and only if** it reduces to a quadratic polynomial (all coefficients of order $> 2$ are zero) and the quadratic part is convex as in (1.2).

---

#### 1.5 Quartic ($d=4$) polynomial

For a quartic polynomial, the Hessian entries are **quadratic** polynomials in $x$. To check convexity, we need:
$$\nabla^2 p(x) \succeq 0 \quad \forall x \in \mathbb{R}^2$$

This requires checking that a matrix whose entries are quadratic polynomials is positive semidefinite everywhere, which is equivalent to checking **non-negativity of polynomials** — a problem that is **NP-hard** in general.

**Key difference from the quadratic case:** For $d=2$, the Hessian is a **constant** matrix, so convexity reduces to a simple eigenvalue/determinant check. For $d=4$, the Hessian depends on $x$, making the check fundamentally harder.

**Example in 1 variable:**
- **Convex quartic:** $f(x) = x^4$. We have $f''(x) = 12x^2 \geq 0$ for all $x$. ✓
- **Non-convex quartic:** $f(x) = -x^4 + 5x^2$. We have $f''(x) = -12x^2 + 10$, which is negative for $|x| > \sqrt{10/12}$, so $f$ is **not convex** (nor concave globally).

## Question 4 (Geometric programming: power control in wireless network) (Mark: 3 + 7 + 6 + 4 = 20 points)

### Example: Power Control in Wireless Communication

 *Adapted from Boyd, Kim, Vandenberghe, and Hassibi,* "[A Tutorial on Geometric Programming](https://web.stanford.edu/~boyd/papers/pdf/gp_tutorial.pdf)."

The [power control problem in wireless communications](http://winlab.rutgers.edu/~narayan/PAPERS/PC%20for%20Wireless%20Data.pdf) aims to minimise the total transmitter power available across $N$ trasmitters while concurrently achieving good (or a pre-defined minimum) performance.

The technical setup is as follows. Each transmitter $i$ transmits with a power level $P_i$ bounded below and above by a minimum and maximum level. The power of the signal received from transmitter $j$ at receiver $i$ is $G_{ij} P_{j}$, where $G_{ij} > 0$ represents the path gain (often loss) from transmitter $j$ to receiver $i$. The signal power at the intended receiver $i$ is $G_{ii} P_i$, and the interference power at receiver $i$ from other transmitters is given by $\sum_{k \neq i} G_{ik}P_k$. The (background) noise power at receiver $i$ is $\sigma_i$. Thus, the _Signal to Interference and  Noise Ratio (SINR)_ of the $i$th receiver-transmitter pair is

$$ S_i = \frac{G_{ii}P_i}{\sum_{k \neq i} G_{ik}P_k + \sigma_i }. $$

The minimum SINR represents a performance lower bound for this system, $S^{\text min}$.

The resulting optimisation problem is formulated as

$$
\begin{array}{ll}
\min_{P} & \sum_{i=1}^N P_i \\
\text{subject to} & P^{min} \leq P_i \leq P^{max}, \; \forall i \\
& \dfrac{G_{ii}P_i}{\sigma_i + \sum_{k \neq i} G_{ik}P_k} \geq S^{min} , \; \forall i \\
\end{array}
$$

### Answer to Question 4

---

#### 4.1 Convert to Standard Geometric Programming Form

The original problem is:
$$\min_{P} \sum_{i=1}^N P_i \quad \text{s.t.} \quad P^{min} \leq P_i \leq P^{max}, \; \frac{G_{ii}P_i}{\sigma_i + \sum_{k\neq i} G_{ik}P_k} \geq S^{min}, \; \forall i$$

**Standard GP form** requires: minimize a posynomial, subject to posynomial $\leq 1$ and monomial $= 1$ constraints.

**Step 1:** The objective $\sum_{i=1}^N P_i$ is a posynomial in $P$. ✓

**Step 2:** Bound constraints:
$$P_i \geq P^{min} \implies P^{min} P_i^{-1} \leq 1$$
$$P_i \leq P^{max} \implies P_i (P^{max})^{-1} \leq 1$$

These are monomial $\leq 1$ constraints. ✓

**Step 3:** SINR constraint — rearrange:
$$G_{ii}P_i \geq S^{min}\left(\sigma_i + \sum_{k\neq i} G_{ik}P_k\right)$$
$$1 \geq \frac{S^{min}\sigma_i}{G_{ii}P_i} + \sum_{k\neq i} \frac{S^{min}G_{ik}}{G_{ii}} P_k P_i^{-1}$$
$$\frac{S^{min}\sigma_i}{G_{ii}} P_i^{-1} + \sum_{k\neq i} \frac{S^{min}G_{ik}}{G_{ii}} P_k P_i^{-1} \leq 1$$

This is a posynomial $\leq 1$ constraint. ✓

Therefore the problem is a valid **Geometric Program** in standard form.

---

#### 4.4 Modified SINR for $N=3$

When the SINR is:
$$S_i = \frac{G_{ii}P_i}{0.5 \sum_{j\neq i} \left(G_{ij}P_j \sum_{k\neq i, k\neq j} G_{ik}P_k\right) + \sigma_i}$$

For $N=3$, the interference term for receiver $i$ involves **products** $P_j \cdot P_k$ (where $j \neq i, k \neq i, k \neq j$). Rearranging the SINR constraint:

$$G_{ii}P_i \geq S^{min}\left(\sigma_i + 0.5 \sum_{j\neq i} G_{ij}P_j \sum_{k\neq i, k\neq j} G_{ik}P_k\right)$$

$$\frac{S^{min}\sigma_i}{G_{ii}} P_i^{-1} + \sum_{j\neq i} \sum_{k\neq i, k\neq j} \frac{0.5 \cdot S^{min} G_{ij} G_{ik}}{G_{ii}} P_j P_k P_i^{-1} \leq 1$$

Each term in the sum has the form $c \cdot P_j P_k P_i^{-1}$, which is a **monomial** with positive coefficient (since all $G_{ij} > 0$). The entire left-hand side is a **sum of monomials with positive coefficients**, i.e., a **posynomial**.

Therefore, this constraint is still of the form posynomial $\leq 1$, and the problem remains a valid **Geometric Program**.

Note: this problem **cannot** be converted to a Linear Program (unlike 4.1 which can), because the interference terms are products of decision variables.

```python
# Q4.2 & Q4.3 — Solve the wireless power control GP using cvxpy
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

np.random.seed(2026)

N = 10
P_min = 0.1
P_max = 5.0
sigma = 0.2  # same for all receivers

# Create random path loss matrix G
# Off-diagonal: uniform in [0.1, 0.9], diagonal = 1
G = np.random.uniform(0.1, 0.9, size=(N, N))
np.fill_diagonal(G, 1.0)

def solve_power_control(N, G, sigma, P_min, P_max, S_min_val):
    """Solve the power control GP for a given S_min."""
    P = cp.Variable(N, pos=True)  # positive variables for GP

    objective = cp.Minimize(cp.sum(P))

    constraints = []
    for i in range(N):
        # Power bounds
        constraints.append(P[i] >= P_min)
        constraints.append(P[i] <= P_max)

        # SINR constraint: G_ii * P_i / (sigma + sum_{k!=i} G_ik * P_k) >= S_min
        if S_min_val > 0:
            interference = sigma
            for k in range(N):
                if k != i:
                    interference += G[i, k] * P[k]
            constraints.append(G[i, i] * P[i] >= S_min_val * interference)

    prob = cp.Problem(objective, constraints)
    try:
        prob.solve(gp=True)
        return prob.status, P.value
    except cp.SolverError:
        return "infeasible", None

# ---- Q4.2: Solve with S_min = 0 ----
print("=" * 60)
print("Q4.2: S_min = 0")
print("=" * 60)
status, P_opt = solve_power_control(N, G, sigma, P_min, P_max, S_min_val=0)
print(f"Status: {status}")
print(f"Optimal power levels: {np.round(P_opt, 4)}")
print(f"Total power: {np.sum(P_opt):.4f}")

# Compute SINRs
sinr = np.zeros(N)
for i in range(N):
    interference = sigma + sum(G[i, k] * P_opt[k] for k in range(N) if k != i)
    sinr[i] = G[i, i] * P_opt[i] / interference
print(f"SINRs: {np.round(sinr, 4)}")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(1, N+1), P_opt, color='steelblue')
axes[0].set_xlabel('Transmitter $i$')
axes[0].set_ylabel('Power $P_i$')
axes[0].set_title('Optimal Power Levels ($S^{min}=0$)')
axes[0].set_xticks(range(1, N+1))

axes[1].bar(range(1, N+1), sinr, color='coral')
axes[1].set_xlabel('Transmitter-Receiver pair $i$')
axes[1].set_ylabel('SINR $S_i$')
axes[1].set_title('SINRs at Optimal Solution ($S^{min}=0$)')
axes[1].set_xticks(range(1, N+1))

plt.tight_layout()
plt.show()

# ---- Q4.3: Solve with increasing S_min ----
print("\n" + "=" * 60)
print("Q4.3: Increasing S_min")
print("=" * 60)

S_min_values = [0.05, 0.1, 0.12, 0.15, 0.2, 0.3, 0.5, 1.0, 5.0]

for S_min_val in S_min_values:
    status, P_opt_s = solve_power_control(N, G, sigma, P_min, P_max, S_min_val)
    if P_opt_s is not None:
        sinr_s = np.zeros(N)
        for i in range(N):
            interference = sigma + sum(G[i, k] * P_opt_s[k] for k in range(N) if k != i)
            sinr_s[i] = G[i, i] * P_opt_s[i] / interference
        print(f"S_min = {S_min_val:5.1f} | Status: {status:10s} | Total power: {np.sum(P_opt_s):8.4f} | Min SINR: {np.min(sinr_s):.4f}")
    else:
        print(f"S_min = {S_min_val:5.1f} | Status: {status}")

print("\nComment: As S_min increases, the total power increases to meet")
print("the stricter SINR requirements. When S_min is too large, the")
print("problem becomes infeasible — no power allocation within [P_min, P_max]")
print("can satisfy all SINR constraints simultaneously.")
```

## Question 5 (Gradient descent for Logistic Regression) (Mark: 5 + 10 + 5 = 20 points)



### Background
Suppose we are given a binary classification dataset consisting of $n$ independent samples $S=\{(\mathbf{x}_i,y_i)\}^n_{i=1}$, where $\mathbf x_i∈\mathbb R^d$ represents the feature vector of the $i$-th sample and $y_i∈\{0,1\}$ is the corresponding label.


Logistic regression models the probability of the event $Y=1$ given the feature vector $\mathbf x$. Specifically, logistic regression assumes a linear relationship between the features $\mathbf{x}$ and the log-odds of the event $Y = 1$:

$$
\log \frac{p(y = 1|\mathbf{x})}{1 - p(y = 1|\mathbf{x})} = \mathbf{x}^\top \mathbf{w},
$$
where $\mathbf w\in \mathbb R^d$ is the weight vector.

Given the dataset, we estimate the weight vector $\mathbf w$ by minimizing the empirical risk under the log-loss (also known as cross-entropy loss):

$$
\hat{R}(\mathbf{w}) = - \frac{1}{n} \sum_{i = 1}^{n} \ell_\mathrm{log}(y_i, \mu_i(\mathbf{w})),
$$

where $\ell_\mathrm{log}(y, \mu) = y \log \mu + (1 - y) \log (1 - \mu)$.

In this exercise you will:

* derive the gradient,
* implement gradient descent from scratch,
* verify correctness by comparing with a standard library implementation.

---





### 5.1 — Mathematical Derivation

$$
\nabla_{\mathbf{w}} \hat{R}(\mathbf{w}_t)=  ?
$$

---

**3.** Using a fixed learning rate ($\eta>0$), write the gradient descent update rule at iteration $t+1$

### Answer to 5.1

---

**1.** From the logistic regression model:
$$\log \frac{p(y=1|\mathbf{x})}{1-p(y=1|\mathbf{x})} = \mathbf{x}^\top \mathbf{w}$$

Solving for $p(y=1|\mathbf{x})$:

$$\mu_i(\mathbf{w}) = p(y_i = 1 | \mathbf{x}_i) = \frac{1}{1 + e^{-\mathbf{x}_i^\top \mathbf{w}}} = \sigma(\mathbf{x}_i^\top \mathbf{w})$$

where $\sigma(z) = \frac{1}{1+e^{-z}}$ is the sigmoid function.

---

**2.** The empirical risk is:
$$\hat{R}(\mathbf{w}) = -\frac{1}{n}\sum_{i=1}^n \left[y_i \log \mu_i(\mathbf{w}) + (1-y_i)\log(1-\mu_i(\mathbf{w}))\right]$$

Using the fact that $\frac{\partial \mu_i}{\partial \mathbf{w}} = \mu_i(1-\mu_i)\mathbf{x}_i$ and the chain rule:

$$\frac{\partial}{\partial \mathbf{w}}\left[y_i \log \mu_i + (1-y_i)\log(1-\mu_i)\right] = y_i(1-\mu_i)\mathbf{x}_i - (1-y_i)\mu_i \mathbf{x}_i = (y_i - \mu_i)\mathbf{x}_i$$

Therefore the gradient is:

$$\boxed{\nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t) = -\frac{1}{n}\sum_{i=1}^n (y_i - \mu_i(\mathbf{w}_t))\mathbf{x}_i = \frac{1}{n}\sum_{i=1}^n (\mu_i(\mathbf{w}_t) - y_i)\mathbf{x}_i}$$

which can be written in matrix form as:
$$\nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t) = \frac{1}{n}\mathbf{X}^\top(\boldsymbol{\mu}(\mathbf{w}_t) - \mathbf{y})$$

where $\mathbf{X} \in \mathbb{R}^{n \times d}$ and $\boldsymbol{\mu} \in \mathbb{R}^n$ with $\mu_i = \sigma(\mathbf{x}_i^\top \mathbf{w}_t)$.

---

**3.** The gradient descent update rule with fixed learning rate $\eta > 0$:

$$\boxed{\mathbf{w}_{t+1} = \mathbf{w}_t - \eta \nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t) = \mathbf{w}_t - \frac{\eta}{n}\mathbf{X}^\top(\boldsymbol{\mu}(\mathbf{w}_t) - \mathbf{y})}$$

where:
- $\mathbf{w}_t \in \mathbb{R}^d$ is the weight vector at iteration $t$
- $\eta > 0$ is the learning rate
- $\boldsymbol{\mu}(\mathbf{w}_t) = \sigma(\mathbf{X}\mathbf{w}_t) \in \mathbb{R}^n$ is the vector of predicted probabilities
- $\mathbf{y} \in \mathbb{R}^n$ is the vector of true labels



### 5.2 — Programming Implementation

```python
#function for generating synthetic data, set your student ID as the random seed. (keep it)
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
plt.rcParams['figure.dpi'] = 108


X, y = make_classification(n_samples = 20, n_features = 2, n_informative=2, n_redundant=0, random_state=2026)
X_b = np.column_stack((np.ones_like(y), X))

plt.scatter(X[y==0,0], X[y==0,1], color='b', label="$y = 0$")
plt.scatter(X[y==1,0], X[y==1,1], color='r', label="$y = 1$")
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```

```python
# Q5.2 — Logistic Regression Implementation
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """Numerically stable sigmoid function."""
    return np.where(z >= 0,
                    1 / (1 + np.exp(-z)),
                    np.exp(z) / (1 + np.exp(z)))

def risk(X, y, w):
    """Compute empirical risk (negative log-likelihood) under log-loss.

    Parameters
    ----------
    X : array of shape (n, d)
    y : array of shape (n,)
    w : array of shape (d,)

    Returns
    -------
    float : empirical risk
    """
    n = X.shape[0]
    mu = sigmoid(X @ w)
    # Clip to avoid log(0)
    eps = 1e-15
    mu = np.clip(mu, eps, 1 - eps)
    return -np.mean(y * np.log(mu) + (1 - y) * np.log(1 - mu))

def grad_risk(X, y, w):
    """Compute the gradient of the empirical risk.

    Parameters
    ----------
    X : array of shape (n, d)
    y : array of shape (n,)
    w : array of shape (d,)

    Returns
    -------
    array of shape (d,) : gradient
    """
    n = X.shape[0]
    mu = sigmoid(X @ w)
    return (1 / n) * X.T @ (mu - y)

def update_weight_gd(X, y, w, lr=1.0, **kwargs):
    """Gradient descent weight update.

    Parameters
    ----------
    X : array of shape (n, d)
    y : array of shape (n,)
    w : array of shape (d,)
    lr : float, learning rate

    Returns
    -------
    array of shape (d,) : updated weight vector
    """
    return w - lr * grad_risk(X, y, w)

def fit_logistic(X, y, w_init=None, max_iter=100, tol=1e-9,
                 update_weight=update_weight_gd, **kwargs):
    """Fits a binary logistic regression model according to the given training
    data.

    Parameters
    ----------
    X : array of shape (n_samples, n_features)
    y : array of shape (n_samples,)
    w_init : array of shape (n_features,), default=None
    max_iter : int, default=100
    tol : float, default=1e-9
    update_weight : callable, default=update_weight_gd
    **kwargs : passed to update_weight

    Returns
    -------
    w_history : list of arrays of shape (n_features,)
    """
    d = X.shape[1]
    if w_init is None:
        w = np.zeros(d)
    else:
        w = w_init.copy()

    w_history = [w.copy()]

    for t in range(max_iter):
        g = grad_risk(X, y, w)
        if np.max(np.abs(g)) <= tol:
            print(f"Converged at iteration {t}, ||grad||_inf = {np.max(np.abs(g)):.2e}")
            break
        w = update_weight(X, y, w, **kwargs)
        w_history.append(w.copy())

    print(f"Final iteration: {len(w_history)-1}, risk = {risk(X, y, w):.6f}, ||grad||_inf = {np.max(np.abs(grad_risk(X, y, w))):.2e}")
    return w_history

# --- Fit the model ---
w_history = fit_logistic(X_b, y, max_iter=10000, tol=1e-9, lr=1.0)
w_final = w_history[-1]
print(f"\nLearned weights: {w_final}")

# --- Visualization: decision boundary ---
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], color='b', label="$y = 0$", edgecolors='k', s=60)
plt.scatter(X[y==1, 0], X[y==1, 1], color='r', label="$y = 1$", edgecolors='k', s=60)

# Decision boundary: w0 + w1*x1 + w2*x2 = 0  =>  x2 = -(w0 + w1*x1) / w2
x1_range = np.linspace(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5, 200)
if np.abs(w_final[2]) > 1e-10:
    x2_boundary = -(w_final[0] + w_final[1] * x1_range) / w_final[2]
    plt.plot(x1_range, x2_boundary, 'g-', linewidth=2, label='Decision boundary')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Logistic Regression — Decision Boundary')
plt.legend()
plt.xlim(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5)
plt.ylim(X[:, 1].min() - 0.5, X[:, 1].max() + 0.5)
plt.show()
```
