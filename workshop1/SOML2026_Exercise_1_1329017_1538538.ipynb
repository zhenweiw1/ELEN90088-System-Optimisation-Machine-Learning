## Question 1 (decide if a polynomial is convex) (Mark: 3 + 5 + 6 + 3 + 3 = 20 points)




Consider a polynomial with $2$ variable
$$
p(x)=\sum_{i,j: i+j\leq d} c_{ij}x_1^{i}x_2^{j}
$$ where the sum is over all nonnegative integer pairs $i,j$ whose sum is less or equal to $d$. We call $d$ the *degree* of the polynomial.

To understand the notation better, consider a polynomial of degree $2$ given by $ h(x)= x_1x_2+2x_2^2+3x_1-1$. You should convince yourself that $h(x)$  can be obtained by setting $c_{11}=1, c_{02}=2, c_{10}=3, c_{00}=-1$ and all other coefficients $c_{ij}$ to $0$ in $p(x)$.

### Answer the following questions:

1. how to decide the convexity of a  first-order ($d=1$) polynomial?
2. how to decide the convexity of a  quadratic ($d=2$) polynomial?
3. how to decide the convexity of a cubic ($d=3$) polynomial?
4. extend the result in (3) to all odd-order ($d$ is odd) polynomial.
5. how to decide the convexity of a quartic ($d=4$) polynomial? What is the difference/difficulty compared to the quadratic case? In particular, try to plot a convex quartic polynomial and a concave quartic polynomial in $1$ variable.

*Hint: It can be helpful to write down the general expression of polynomials p(x) before deciding the convexity.*

*Comment: as shown in [this paper](https://web.mit.edu/~a_a_a/Public/Publications/convexity_nphard.pdf), unless NP=P, it is computationally hard to check the convexity of a polynomial with order equal or larger than $4$. So in general, it is computationaly difficult to decide if a function is convex.*

### Answer to Question 1

---

#### 1.1 First-order ($d=1$) polynomial

A first-order polynomial can be written as
$$p(x)=c_{10}x_1+c_{01}x_2+c_{00}.$$

It is an affine function, so its Hessian is always zero:
$$\nabla^2 p(x)=\mathbf 0 \succeq 0.$$

Therefore, every first-order polynomial is both convex and concave.

---

#### 1.2 Quadratic ($d=2$) polynomial

A quadratic polynomial is
$$p(x)=c_{20}x_1^2+c_{11}x_1x_2+c_{02}x_2^2+c_{10}x_1+c_{01}x_2+c_{00}.$$

Its Hessian is the constant matrix
$$\nabla^2 p(x)=\begin{bmatrix}2c_{20}&c_{11}\\c_{11}&2c_{02}\end{bmatrix}.$$

So we only need to check whether this matrix is positive semidefinite. The conditions are
1. $c_{20}\ge 0$ and $c_{02}\ge 0$
2. $4c_{20}c_{02}-c_{11}^2\ge 0$

---

#### 1.3 Cubic ($d=3$) polynomial

For a cubic polynomial, each entry of the Hessian is linear in $x$. If the Hessian must be positive semidefinite for all $x$, those linear terms cannot stay nonnegative everywhere unless all cubic coefficients are zero.

So a cubic polynomial can be convex only if it first reduces to a quadratic polynomial, and then the quadratic part must satisfy the conditions in 1.2.

---

#### 1.4 All odd-order ($d$ odd) polynomials

The same idea extends to every odd degree. The highest-order part of the Hessian is still an odd-degree polynomial, so it changes sign as $x\to\pm\infty$. That means it cannot stay positive semidefinite everywhere.

Therefore, if an odd-degree polynomial is convex, it must actually reduce to a quadratic polynomial and then satisfy the quadratic convexity conditions.

---

#### 1.5 Quartic ($d=4$) polynomial

For a quartic polynomial, the Hessian is no longer constant. Its entries are quadratic polynomials in $x$, so checking convexity is much harder than in the quadratic case.

- Quadratic case: check whether one constant matrix is positive semidefinite
- Quartic case: check whether the Hessian is positive semidefinite for every $x$

That is why convexity checking becomes much harder for quartic and higher-order polynomials.

Examples in one variable:
- convex quartic: $f(x)=x^4$, since $f''(x)=12x^2\ge 0$
- concave quartic: $f(x)=-x^4$, since $f''(x)=-12x^2\le 0$

See the plot below.

```python
# Q1.5 - Plot convex and concave quartic polynomials in 1 variable
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 500)

f_convex = x**4
f_concave = -x**4

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, f_convex, 'b-', linewidth=2)
axes[0].set_title('Convex quartic: $f(x) = x^4$')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$f(x)$')
axes[0].grid(True)

axes[1].plot(x, f_concave, 'r-', linewidth=2)
axes[1].set_title('Concave quartic: $f(x) = -x^4$')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$f(x)$')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### 2.1

$$E_{total}=E_{comp}+E_{tx}=c_1f^2+c_2t_{tx}^{-1}$$

### 2.2

From the assumption that the optimum satisfies

The plotting code is shown below.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define physical constants
c1 = 1  # Hardware capacitance constant
c2 = 2  # Channel-dependent constant
c3 = 1  # CPU cycles for the task
T  = 3  # Latency deadline (seconds)

# 2. Determine the valid domain
f_min = c3 / T
f = np.linspace(f_min + 0.01, 3.0, 500)

# 3. Calculate the Energy Function E(f)
energy = c1 * (f**2) + (c2 * f) / (f * T - c3)

# 4. Create the Plot
plt.figure(figsize=(10, 6))
plt.plot(f, energy, label=r'$E(f) = c_1 f^2 + \frac{c_2 f}{f T - c_3}$',
         color='#1f77b4', linewidth=2.5)

plt.axvline(x=f_min, color='red', linestyle='--', alpha=0.6,
            label=f'Domain Boundary ($f > {f_min:.4f}$)')

# 5. Formatting the chart
plt.title('Edge Device Energy Optimization', fontsize=14, fontweight='bold')
plt.xlabel('CPU Frequency $f$ (Hz)', fontsize=12)
plt.ylabel('Total Energy Consumption $E(f)$ (Joules)', fontsize=12)

# Limit Y-axis to prevent the asymptote from squashing the rest of the curve
plt.ylim(0, np.percentile(energy, 95))

plt.grid(True, which='both', linestyle=':', alpha=0.7)
plt.legend(fontsize=11)

plt.tight_layout()
plt.show()
```

### 2.3

The code is as below,

```python
import cvxpy as cp

# 1. Define the physical constants
c1 = 1e-18   # Hardware capacitance
c2 = 0.1     # Channel-dependent constant
c3 = 10**9   # CPU cycles required
T  = 2.0     # Latency deadline (seconds)

# 2. Define the Optimization Variables
f = cp.Variable(pos=True)
t_tx = cp.Variable(pos=True)

# 3. Define the Objective Function
objective_fn = c1 * cp.power(f, 2) + c2 * cp.inv_pos(t_tx)
objective = cp.Minimize(objective_fn)

# 4. Define the Constraints
constraints = [(c3 * cp.inv_pos(f) + t_tx) / T <= 1]

# 5. Solve the Problem
prob = cp.Problem(objective, constraints)
prob.solve(gp=True)

# 6. Output the Results
if prob.status == cp.OPTIMAL:
    print("="*40)
    print("OPTIMAL SOLUTION FOUND")
    print("="*40)
    print(f"Optimal CPU Frequency (f*):      {f.value:.4f} Hz")
    print(f"Optimal Trans. Time (t_tx*):     {t_tx.value:.4f} seconds")
    print(f"Computation Time (t_comp):      {(c3/f.value):.4f} seconds")
    print(f"Minimum Total Energy:           {prob.value:.8f} Joules")
    print("="*40)
else:
    print("Solver could not find an optimal solution.")
```

## Question 3 (Aloha communication protocol) (Mark: 3 + 15 + 2 = 20 points)

**Aloha** is a well-known random access or _MAC_ (Media/multiple Access Control) communication protocol. In this system, multiple independent transmitters send packets (called $\textit{frames}$) to a single receiver over a shared channel. In the original Aloha, each transmitter sends whenever it has data, without coordinating with others. If two or more packets overlap in time, a collision occurs and the packets are lost; the transmitters then retry after a random delay.

In a more advanced setup, called **slotted Aloha**, the nodes can only transmit at the beginning of time slots, which are kept by a global/shared clock. The operations of slotted ALOHA in each node are decribed as follows:

* If there is only one frame (namely only one active transmitter) in a slot, then the transmission is successful and the slot is said to be a successful slot.
* If two or more frames (namely two or more active transmitters) collide in a slot, then the transmission is failed and a re-transmission is considered for each frame involved in the collision. The transmitter will retransmit its frame in each subsequent slot with probability $p$ until the frame is transmitted without a collision, where $0 \leq p \leq 1$.

Assume that there are $N$ transmitters and each transmitter independently attempts to transmit a frame in each slot with probability $p$. For one slot, let $S$ denote the _success probability_ that this slot is a successful slot.

Answer the following questions:
1. Provide an expression of $S$ defined above.
2. To maximise the $S$, define the optimisation problem to find the optimal value $p$. Note that the constraint $0 \leq p \leq 1$
should be taken into consideration. Clearly identify the objective and decision variable(s). Is the objective convex or concave? Show through derivation. Find the optimality conditions for this problem.
3. When $N$ tends to infinity, what is the maximal success probability $S$?

## Answer:

### 3.1

A slot is successful only when exactly one transmitter sends and all others stay silent. Therefore,

$$S(p)=Np(1-p)^{N-1}.$$

### 3.2

The optimization problem is

$$\max_{0\le p\le 1} S(p)=Np(1-p)^{N-1}.$$

- Decision variable: $p$
- Objective: maximize $S(p)$

First derivative:

$$
\begin{aligned}
S'(p)
&=\frac{d}{dp}[Np(1-p)^{N-1}]\\
&=N(1-p)^{N-2}(1-Np).
\end{aligned}
$$

So the stationary point satisfies
$$S'(p)=0 \Longrightarrow p^*=\frac{1}{N}.$$

Second derivative:

$$
\begin{aligned}
S''(p)
&=N(N-1)(1-p)^{N-3}(Np-2).
\end{aligned}
$$

For $N\ge 3$:
- if $p<2/N$, then $S''(p)<0$
- if $p>2/N$, then $S''(p)>0$

So $S(p)$ is neither globally convex nor globally concave on $[0,1]$.

At the boundary,
$$S(0)=0,\qquad S(1)=0,$$
while
$$S(1/N)=\left(1-\frac{1}{N}\right)^{N-1}>0.$$

Hence the global maximizer is
$$p^*=\frac{1}{N}.$$

### 3.3

Substituting $p^*=1/N$ gives

$$S^*=\left(1-\frac{1}{N}\right)^{N-1}.$$

As $N\to\infty$,

$$\lim_{N\to\infty}S^*=\frac{1}{e}\approx 0.368.$$

## Question 4 (Geometric programming: power control in wireless network) (Mark: 3 + 7 + 6 + 4 = 20 points)

### Example: Power Control in Wireless Communication

 *Adapted from Boyd, Kim, Vandenberghe, and Hassibi,* "[A Tutorial on Geometric Programming](https://web.stanford.edu/~boyd/papers/pdf/gp_tutorial.pdf)."

The [power control problem in wireless communications](http://winlab.rutgers.edu/~narayan/PAPERS/PC%20for%20Wireless%20Data.pdf) aims to minimise the total transmitter power available across $N$ trasmitters while concurrently achieving good (or a pre-defined minimum) performance.

The technical setup is as follows. Each transmitter $i$ transmits with a power level $P_i$ bounded below and above by a minimum and maximum level. The power of the signal received from transmitter $j$ at receiver $i$ is $G_{ij} P_{j}$, where $G_{ij} > 0$ represents the path gain (often loss) from transmitter $j$ to receiver $i$. The signal power at the intended receiver $i$ is $G_{ii} P_i$, and the interference power at receiver $i$ from other transmitters is given by $\sum_{k \neq i} G_{ik}P_k$. The (background) noise power at receiver $i$ is $\sigma_i$. Thus, the _Signal to Interference and  Noise Ratio (SINR)_ of the $i$th receiver-transmitter pair is

$$ S_i = \frac{G_{ii}P_i}{\sum_{k \neq i} G_{ik}P_k + \sigma_i }. $$

The minimum SINR represents a performance lower bound for this system, $S^{\text min}$.

The resulting optimisation problem is formulated as

$$
\begin{array}{ll}
\min_{P} & \sum_{i=1}^N P_i \\
\text{subject to} & P^{min} \leq P_i \leq P^{max}, \; \forall i \\
& \dfrac{G_{ii}P_i}{\sigma_i + \sum_{k \neq i} G_{ik}P_k} \geq S^{min} , \; \forall i \\
\end{array}
$$

### Answer the following questions: Wireless Power Control

Let $N=10$, $P^{min}=0.1$, $P^{max}=5$, $\sigma=0.2$ (same for all). Create a random path loss matrix $G$, where off-diagonal elements are between $0.1$ and $0.9$ and the diagonal elements are equal to $1$.

### Answer to Question 4

#### 4.1 Show that the problem is a geometric program

The original problem is
$$
\min_{P} \sum_{i=1}^{N} P_i
$$
subject to
$$
P^{\min} \le P_i \le P^{\max},
$$
and
$$
\frac{G_{ii} P_i}{\sigma_i + \sum_{k \ne i} G_{ik} P_k} \ge S^{\min}.
$$

To write it in GP form:

1. The objective
$$
\sum_{i=1}^{N} P_i
$$
is a posynomial.

2. The power bounds become
$$
P_i \ge P^{\min} \quad \Longrightarrow \quad P^{\min} P_i^{-1} \le 1,
$$
and
$$
P_i \le P^{\max} \quad \Longrightarrow \quad P_i (P^{\max})^{-1} \le 1.
$$
These are monomial constraints.

3. The SINR constraint becomes
$$
G_{ii} P_i \ge S^{\min} \left( \sigma_i + \sum_{k \ne i} G_{ik} P_k \right).
$$
Dividing both sides by $G_{ii} P_i$ gives
$$
\frac{S^{\min} \sigma_i}{G_{ii}} P_i^{-1}
+ \sum_{k \ne i} \frac{S^{\min} G_{ik}}{G_{ii}} P_k P_i^{-1}
\le 1.
$$
The left-hand side is a posynomial.

So the problem is a standard GP.

---

#### 4.2 Solve the problem with $S^{\min} = 0$

When $S^{\min} = 0$, the SINR constraints disappear, so the problem becomes
$$
\min_{P} \sum_{i=1}^{N} P_i
\quad \text{subject to} \quad
P^{\min} \le P_i \le P^{\max}.
$$

To minimize total power, every transmitter should use the smallest allowed power:
$$
P_i^* = P^{\min} = 0.1, \quad i = 1, \dots, N.
$$

Since $N = 10$, the optimal total power is
$$
\sum_{i=1}^{N} P_i^* = 10 \times 0.1 = 1.0.
$$

---

#### 4.3 What happens when $S^{\min}$ increases?

As $S^{\min}$ increases, the SINR constraints become stricter:
- for small $S^{\min}$, low power may still be feasible
- for larger $S^{\min}$, the total power must increase
- if $S^{\min}$ is too large, the problem becomes infeasible

So the trend is
$$
\text{larger } S^{\min}
\quad \Longrightarrow \quad
\text{larger total power, and eventually infeasibility}.
$$

---

#### 4.4 Modified SINR for $N = 3$

The modified SINR is
$$
S_i =
\frac{G_{ii} P_i}{0.5 \sum_{j \ne i} \left( G_{ij} P_j \sum_{k \ne i,\, k \ne j} G_{ik} P_k \right) + \sigma_i}.
$$

Imposing $S_i \ge S^{\min}$ gives
$$
G_{ii} P_i
\ge
S^{\min}
\left[
0.5 \sum_{j \ne i} \left( G_{ij} P_j \sum_{k \ne i,\, k \ne j} G_{ik} P_k \right) + \sigma_i
\right].
$$

Dividing by $G_{ii} P_i$ gives
$$
\frac{S^{\min} \sigma_i}{G_{ii}} P_i^{-1}
+ \sum_{j \ne i} \sum_{k \ne i,\, k \ne j}
\frac{0.5 S^{\min} G_{ij} G_{ik}}{G_{ii}} P_j P_k P_i^{-1}
\le 1.
$$

Each term is a monomial, so the whole left-hand side is a posynomial. Therefore the problem is still a GP.

However, because products like $P_j P_k$ appear, it is not an LP.

```python
# Q4.2 & Q4.3 — Solve the wireless power control GP using cvxpy
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

np.random.seed(2026)

N = 10
P_min = 0.1
P_max = 5.0
sigma = 0.2  # same for all receivers

# Create random path loss matrix G
# Off-diagonal: uniform in [0.1, 0.9], diagonal = 1
G = np.random.uniform(0.1, 0.9, size=(N, N))
np.fill_diagonal(G, 1.0)

def solve_power_control(N, G, sigma, P_min, P_max, S_min_val):
    """Solve the power control GP for a given S_min."""
    P = cp.Variable(N, pos=True)  # positive variables for GP

    objective = cp.Minimize(cp.sum(P))

    constraints = []
    for i in range(N):
        # Power bounds
        constraints.append(P[i] >= P_min)
        constraints.append(P[i] <= P_max)

        # SINR constraint: G_ii * P_i / (sigma + sum_{k!=i} G_ik * P_k) >= S_min
        if S_min_val > 0:
            interference = sigma
            for k in range(N):
                if k != i:
                    interference += G[i, k] * P[k]
            constraints.append(G[i, i] * P[i] >= S_min_val * interference)

    prob = cp.Problem(objective, constraints)
    try:
        prob.solve(gp=True)
        return prob.status, P.value
    except cp.SolverError:
        return "infeasible", None

# ---- Q4.2: Solve with S_min = 0 ----
print("=" * 60)
print("Q4.2: S_min = 0")
print("=" * 60)
status, P_opt = solve_power_control(N, G, sigma, P_min, P_max, S_min_val=0)
print(f"Status: {status}")
print(f"Optimal power levels: {np.round(P_opt, 4)}")
print(f"Total power: {np.sum(P_opt):.4f}")

# Compute SINRs
sinr = np.zeros(N)
for i in range(N):
    interference = sigma + sum(G[i, k] * P_opt[k] for k in range(N) if k != i)
    sinr[i] = G[i, i] * P_opt[i] / interference
print(f"SINRs: {np.round(sinr, 4)}")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(1, N+1), P_opt, color='steelblue')
axes[0].set_xlabel('Transmitter $i$')
axes[0].set_ylabel('Power $P_i$')
axes[0].set_title('Optimal Power Levels ($S^{min}=0$)')
axes[0].set_xticks(range(1, N+1))

axes[1].bar(range(1, N+1), sinr, color='coral')
axes[1].set_xlabel('Transmitter-Receiver pair $i$')
axes[1].set_ylabel('SINR $S_i$')
axes[1].set_title('SINRs at Optimal Solution ($S^{min}=0$)')
axes[1].set_xticks(range(1, N+1))

plt.tight_layout()
plt.show()

# ---- Q4.3: Solve with increasing S_min ----
print("\n" + "=" * 60)
print("Q4.3: Increasing S_min")
print("=" * 60)

S_min_values = [0.05, 0.1, 0.12, 0.15, 0.2, 0.3, 0.5, 1.0, 5.0]

for S_min_val in S_min_values:
    status, P_opt_s = solve_power_control(N, G, sigma, P_min, P_max, S_min_val)
    if P_opt_s is not None:
        sinr_s = np.zeros(N)
        for i in range(N):
            interference = sigma + sum(G[i, k] * P_opt_s[k] for k in range(N) if k != i)
            sinr_s[i] = G[i, i] * P_opt_s[i] / interference
        print(f"S_min = {S_min_val:5.2f} | Status: {status:10s} | Total power: {np.sum(P_opt_s):8.4f} | Min SINR: {np.min(sinr_s):.4f}")
    else:
        print(f"S_min = {S_min_val:5.2f} | Status: {status}")

print("\nComment: As S_min increases, the total power increases to meet")
print("the stricter SINR requirements. When S_min is too large, the")
print("problem becomes infeasible — no power allocation within [P_min, P_max]")
print("can satisfy all SINR constraints simultaneously.")
```

## Question 5 (Gradient descent for Logistic Regression) (Mark: 5 + 10 + 5 = 20 points)



### Background
Suppose we are given a binary classification dataset consisting of $n$ independent samples $S=\{(\mathbf{x}_i,y_i)\}^n_{i=1}$, where $\mathbf x_i∈\mathbb R^d$ represents the feature vector of the $i$-th sample and $y_i∈\{0,1\}$ is the corresponding label.


Logistic regression models the probability of the event $Y=1$ given the feature vector $\mathbf x$. Specifically, logistic regression assumes a linear relationship between the features $\mathbf{x}$ and the log-odds of the event $Y = 1$:

$$
\log \frac{p(y = 1|\mathbf{x})}{1 - p(y = 1|\mathbf{x})} = \mathbf{x}^\top \mathbf{w},
$$
where $\mathbf w\in \mathbb R^d$ is the weight vector.

Given the dataset, we estimate the weight vector $\mathbf w$ by minimizing the empirical risk under the log-loss (also known as cross-entropy loss):

$$
\hat{R}(\mathbf{w}) = - \frac{1}{n} \sum_{i = 1}^{n} \ell_\mathrm{log}(y_i, \mu_i(\mathbf{w})),
$$

where $\ell_\mathrm{log}(y, \mu) = y \log \mu + (1 - y) \log (1 - \mu)$.

In this exercise you will:

* derive the gradient,
* implement gradient descent from scratch,
* verify correctness by comparing with a standard library implementation.

---



### 5.1 — Mathematical Derivation

**1.** Given the expression
$$
\mu_i(\mathbf{w}) := \frac{1}{1 + \exp(-\mathbf{x}_i^\top \mathbf{w})}
$$

**2.** Give the expression of the gradient of the empirical risk at iteration $t$:
$$
\nabla_{\mathbf{w}} \hat{R}(\mathbf{w}_t) = \frac{1}{n} \sum_{i=1}^{n} (\mu_i - y_i) \, \mathbf{x}_i
$$

---

**3.** Using a fixed learning rate ($\eta>0$), write the gradient descent update rule at iteration $t+1$:

$$
\mathbf{w}_{t+1} = \mathbf{w}_t - \frac{\eta}{n} \sum_{i=1}^{n} (\mu_i(\mathbf{w}_t) - y_i) \, \mathbf{x}_i
$$

Clearly specify all quantities appearing in the update, and note that $\mathbf{w}, \mathbf{x}_i \in \mathbb{R}^{d}$.

| Quantity | Dimension |
| :--- | :--- |
| $\mathbf{w}_{t+1}$ | $\mathbb{R}^d$ |
| $\mathbf{w}_t$ | $\mathbb{R}^d$ |
| $\eta$ | $\mathbb{R}_{>0}$ |
| $n$ | $\mathbb{Z}^+$ |
| $y_i$ | $\{0, 1\}$ |
| $\mu_i$ | $(0, 1)$ |
| $\mathbf{x}_i$ | $\mathbb{R}^d$ |
---

### Answer to 5.1

---

**1.** From
$$
\log \frac{p(y=1\mid\mathbf{x})}{1-p(y=1\mid\mathbf{x})}=\mathbf{x}^\top\mathbf{w}
$$
we get
$$
\mu_i(\mathbf{w})=\frac{1}{1+e^{-\mathbf{x}_i^\top\mathbf{w}}}=\sigma(\mathbf{x}_i^\top\mathbf{w})
$$

---

**2.** The empirical risk is
$$
\hat{R}(\mathbf{w})=-\frac{1}{n}\sum_{i=1}^n\left[y_i\log\mu_i+(1-y_i)\log(1-\mu_i)\right]
$$

Since
$$
\frac{\partial \mu_i}{\partial \mathbf{w}}=\mu_i(1-\mu_i)\mathbf{x}_i
$$
the gradient is
$$
\boxed{\nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t)=\frac{1}{n}\sum_{i=1}^n(\mu_i(\mathbf{w}_t)-y_i)\,\mathbf{x}_i}
$$

In matrix form,
$$
\nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t)=\frac{1}{n}\mathbf{X}^\top\bigl(\boldsymbol{\mu}(\mathbf{w}_t)-\mathbf{y}\bigr)
$$

---

**3.** The gradient descent update is
$$
\boxed{\mathbf{w}_{t+1}=\mathbf{w}_t-\eta\nabla_{\mathbf{w}}\hat{R}(\mathbf{w}_t)=\mathbf{w}_t-\frac{\eta}{n}\mathbf{X}^\top\bigl(\boldsymbol{\mu}(\mathbf{w}_t)-\mathbf{y}\bigr)}
$$

where:
- $\mathbf{w}_t$ is the parameter at iteration $t$
- $\eta$ is the learning rate
- $\boldsymbol{\mu}(\mathbf{w}_t)$ is the vector of predicted probabilities
- $\mathbf{y}$ is the label vector



### 5.2 — Programming Implementation

```python
# Q5.2 — Logistic Regression Implementation
import numpy as np
import matplotlib.pyplot as plt

C_REG = 1.0
REGULARIZE_BIAS = False

def sigmoid(z):
    """Numerically stable sigmoid function."""
    return np.where(z >= 0,
                    1 / (1 + np.exp(-z)),
                    np.exp(z) / (1 + np.exp(z)))

def risk(X, y, w, C=C_REG, regularize_bias=REGULARIZE_BIAS):
    """Compute regularized empirical risk under log-loss."""
    n = X.shape[0]
    mu = sigmoid(X @ w)
    eps = 1e-15
    mu = np.clip(mu, eps, 1 - eps)
    data_loss = -np.mean(y * np.log(mu) + (1 - y) * np.log(1 - mu))

    reg_slice = slice(0, None) if regularize_bias else slice(1, None)
    reg = 0.5 * np.sum(w[reg_slice] ** 2) / (C * n)
    return data_loss + reg

def grad_risk(X, y, w, C=C_REG, regularize_bias=REGULARIZE_BIAS):
    """Compute the gradient of the regularized empirical risk."""
    n = X.shape[0]
    mu = sigmoid(X @ w)
    grad = (1 / n) * X.T @ (mu - y)

    reg_grad = np.zeros_like(w)
    reg_slice = slice(0, None) if regularize_bias else slice(1, None)
    reg_grad[reg_slice] = w[reg_slice] / (C * n)
    return grad + reg_grad

def update_weight_gd(X, y, w, lr=1.0, **kwargs):
    """Gradient descent weight update."""
    return w - lr * grad_risk(X, y, w, **kwargs)

def fit_logistic(X, y, w_init=None, max_iter=100, tol=1e-9, lr=0.5,
                 update_weight=update_weight_gd, **kwargs):
    """Fit a regularized binary logistic regression model with batch gradient descent."""
    d = X.shape[1]
    if w_init is None:
        w = np.zeros(d)
    else:
        w = w_init.copy()

    for t in range(max_iter):
        g = grad_risk(X, y, w, **kwargs)
        if np.max(np.abs(g)) <= tol:
            print(f"Converged at iteration {t}, ||grad||_inf = {np.max(np.abs(g)):.2e}")
            break
        w = update_weight(X, y, w, lr=lr, **kwargs)

    print(f"Final iteration: {t+1}, risk = {risk(X, y, w, **kwargs):.6f}, ||grad||_inf = {np.max(np.abs(grad_risk(X, y, w, **kwargs))):.2e}")
    return w

# --- Fit the model ---
w_final = fit_logistic(
    X_b,
    y,
    max_iter=50000,
    tol=1e-9,
    lr=0.5,
    C=C_REG,
    regularize_bias=REGULARIZE_BIAS,
)
print(f"\nUsing L2 regularization with C = {C_REG}")
print(f"Learned weights: {w_final}")

# --- Visualization: decision boundary ---
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], color='b', label="$y = 0$", edgecolors='k', s=60)
plt.scatter(X[y==1, 0], X[y==1, 1], color='r', label="$y = 1$", edgecolors='k', s=60)

# Decision boundary: w0 + w1*x1 + w2*x2 = 0  =>  x2 = -(w0 + w1*x1) / w2
x1_range = np.linspace(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5, 200)
if np.abs(w_final[2]) > 1e-10:
    x2_boundary = -(w_final[0] + w_final[1] * x1_range) / w_final[2]
    plt.plot(x1_range, x2_boundary, 'g-', linewidth=2, label='Decision boundary')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Regularized Logistic Regression — Decision Boundary')
plt.legend()
plt.xlim(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5)
plt.ylim(X[:, 1].min() - 0.5, X[:, 1].max() + 0.5)
plt.show()
```
